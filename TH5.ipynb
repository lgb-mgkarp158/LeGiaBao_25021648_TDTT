```python
#W5A1
def Max_Of_Two(a,b) -> int:
    """Trả về số lớn hơn trong hai số nguyên a và b."""
    if a>=b:
        return a
    else: 
        return b
a,b = map(int, input('Nhập vào hai số nguyên a và b').split())
ketqua= Max_Of_Two(a,b)
print (ketqua)
```

```python
#W5A2
from typing import Tuple

def swap(a, b) -> Tuple[int, int]:
    """Hoán đổi giá trị của hai số nguyên a và b."""
    return b, a
a,b = map(int, input('Nhập vào hai số nguyên a và b').split())
ketqua= swap(a,b)
print (ketqua)   
```

```python
#W5A3
def so_nguyento(n: int) :
    """ Kiểm tra một số nguyên n có phải là số nguyên tố không.  """
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True
n = int(input(' Nhập số n cần kiểm tra :'))
print(so_nguyento(n))
```

```python
#W5A4
def so_hoanhao(n):
    if n <= 0:        
        return False

    total = 0
    for i in range(1, n):
        if n % i == 0:  
            total += i

    return total == n 
n = int(input(' Nhập số n cần kiểm tra :'))
print(so_hoanhao(n))
```

```python
#W5A5
def find_index(lst, x):
    for i in range(len(lst)):     
        if lst[i] == x:         
            return i              
    return -1                   
n = list(map(int, input('Nhập vào danh sách').split()))
x = int(input(' Nhập số cần kiểm tra :'))
print(find_index(n,x))
```

```python
#W5A6
def luythua(n):
    if n < 0:
        return None    

    result = 1
    for i in range(1, n+1):
        result *= i     
    return result
n = int(input(' Nhập số n cần kiểm tra :'))
print(luythua(n))
```

```python
#W5A7
def calc(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b == 0:
            return None
        return a / b
    else:
        return None     
a = int(input('Nhập  a: '))
b = int(input('Nhập b: '))
phep = input('Nhập phép tính +; -; *; /')
kq = calc(a,b,phep)
print (kq)
```

```python
#W5A8
def hamming_distance(x: int, y: int) -> int:
    xor_result = x ^ y
    distance = bin(xor_result).count('1')
    return distance  
num1 = int(input("Nhập số nguyên thứ nhất x: "))
num2 = int(input("Nhập số nguyên thứ hai y: "))
result = hamming_distance(num1, num2) 
print(f"\nKhoảng cách Hamming giữa {num1} và {num2} là: {result}")
```

```python
#W5A9
def sum_digits(n):
    total = 0
    for ch in str(n):
        total += int(ch)
    return total
n = int(input(' Nhập số n cần kiểm tra :'))
print(sum_digits(n))
```

```python
#W5A10
def same_pattern(a, b):
    if len(a) != len(b):
        return False

    mapA = {}
    mapB = {}

    for ch1, ch2 in zip(a, b):
        # Nếu đã có ánh xạ nhưng không khớp
        if (ch1 in mapA and mapA[ch1] != ch2) or \
           (ch2 in mapB and mapB[ch2] != ch1):
            return False

        # Cập nhật ánh xạ
        mapA[ch1] = ch2
        mapB[ch2] = ch1

    return True
word_a = input("Nhập từ thứ nhất (A): ").strip().lower()
word_b = input("Nhập từ thứ hai (B): ").strip().lower()
result = same_pattern(word_a, word_b)
if result:
    print(f"\nKết quả: True. '{word_a}' và '{word_b}' là từ Đảo Cấu.")
else:
    print(f"\nKết quả: False. '{word_a}' và '{word_b}' không phải là từ Đảo Cấu.")
```
