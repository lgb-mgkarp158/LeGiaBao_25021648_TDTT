```python
#W5A1
def Max_Of_Two(a,b) -> int:
    """Trả về số lớn hơn trong hai số nguyên a và b."""
    if a>=b:
        return a
    else: 
        return b
a,b = map(int, input('Nhập vào hai số nguyên a và b').split())
ketqua= Max_Of_Two(a,b)
print (ketqua)
```

```python
#W5A3
def so_nguyento(n: int) :
    """ Kiểm tra một số nguyên n có phải là số nguyên tố không.  """
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True
n = int(input(' Nhập số n cần kiểm tra :'))
print(so_nguyento(n))
```

```python
#W5A4
def so_hoanhao(n):
    if n <= 0:        
        return False

    total = 0
    for i in range(1, n):
        if n % i == 0:  
            total += i

    return total == n 
n = int(input(' Nhập số n cần kiểm tra :'))
print(so_hoanhao(n))
```

```python
#W5A5
def find_index(lst, x):
    for i in range(len(lst)):     
        if lst[i] == x:         
            return i              
    return -1                   
n = list(map(int, input('Nhập vào danh sách').split()))
x = int(input(' Nhập số cần kiểm tra :'))
print(find_index(n,x))
```

```python
#W5A6
def luythua(n):
    if n < 0:
        return None    

    result = 1
    for i in range(1, n+1):
        result *= i     
    return result
n = int(input(' Nhập số n cần kiểm tra :'))
print(luythua(n))
```

```python
#W5A7
def calc(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b == 0:
            return None
        return a / b
    else:
        return None     
a = int(input('Nhập  a: '))
b = int(input('Nhập b: '))
phep = input('Nhập phép tính +; -; *; /')
kq = calc(a,b,phep)
print (kq)
```

```python
#W5A9
def sum_digits(n):
    total = 0
    for ch in str(n):
        total += int(ch)
    return total
n = int(input(' Nhập số n cần kiểm tra :'))
print(sum_digits(n))
```

```python
#W5A10
def dang_cau(a, b):
    so_sanh = {}
    for i in range(len(a)):
        char_a = a[i] 
        char_b = b[i]
        if char_a in so_sanh:
            if so_sanh[char_a] != char_b:
                return False
        else:
            if char_b in so_sanh.values():
                return False
            else:
                so_sanh[char_a] = char_b
    return True
```
