```python
A = int(input('Nhập vào số nguyên dương A: '))
sum = 0
n = 0
while sum <= A:
    n += 1
    sum += 1/n
print('Giá trị nhỏ nhất là', n)
```

```python
max = None
min = None
while True:
    x =  int(input("Nhập số nguyên"))
    if x == -1:
        break
    if max is None or x > max:
        max = x
    if min is None or x< min:
        min = x
print("Số lớn nhất:", max)
print("Số bé nhất:", min)    
```

```python
n = int(input("Nhập số nguyên dương n: "))
chan = 0
le = 0
while n > 0:
    donvi = n % 10
    if donvi % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print("Chữ số chẵn:", chan)
print("Chữ số lẻ:", le)
```

```python
n = int(input("Nhập số nguyên dương n: "))
x = 1
luythua= False
while x<=n:
    if x ==n:
        luythua = True
        break
    x *= 3
print("Là số dạng 3^k" if luythua else "Không phải số dạng 3^k")
```

```python
A = int(input("Nhập số nguyên dương A: "))
f1, f2 = 1, 1
while f2 <= A:
    f1, f2 = f2, f1 + f2
print("Số Fibonacci lớn nhất không vượt quá A là:", f1)
```

```python
n = int(input("Nhập số nguyên dương n: "))
count = 0
while n > 0:
    digit = n % 10
    if digit in [2, 3, 5, 7]:
        count += 1
    n //= 10
print("Số chữ số là số nguyên tố:", count)
```
