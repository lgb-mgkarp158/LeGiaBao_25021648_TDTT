```python
#W4A4
n = abs(int(input("Nhập số nguyên: ")))
count = 0
while n > 0:
    count += 1
    n //= 10
print("Số chữ số là:", count)
```

```python
#W4A6
a, b = map(int, input("Nhập hai số nguyên dương a và b (a ≤ b): ").split())
sum = 0
for i in range(a,b + 1):
    if i <2:
        continue
    nguyen_to = True
    for n in range (2, int(n**0.5)+1):
        if i%n ==0:
            nguyen_to = False
            break
    if nguyen_to:
        sum += i
print("Tổng các số nguyên tố trong đoạn từ", a, "đến", b, "là:", sum)
```

```python
#W4A7
n = int(input())

i = 2
lon_nhat = 1
while i * i <= n:
    while n % i == 0:
        lon_nhat = i
        n //= i
    i += 1

if n > 1:
    lon_nhat = n

print(lon_nhat)
```
