```python
#W4A6
a, b = map(int, input("Nhập hai số nguyên dương a và b (a ≤ b): ").split())
sum = 0

for i in range(a,b + 1):
    if i <2:
        continue
    nguyen_to = True
    for n in range (2, int(i**0.5)+1):
        if i%n ==0:
            nguyen_to = False
            break
    if nguyen_to:
        sum += i
print("Tổng các số nguyên tố trong đoạn từ", a, "đến", b, "là:", sum)
```

```python
#W4A8
try:
    n = int(input("Nhập số nguyên dương n: "))
    if n <= 0:
        print("Lỗi: Vui lòng nhập số nguyên dương.")
    else:
        buoc = 0
        current_n = n
        while True:
            def dao_nguoc(x):
                rev = 0
                while x > 0:   
                    rev = rev * 10 + x % 10
                    x = x // 10
                return rev

            if current_n == dao_nguoc(current_n):
                break

            current_n = current_n + dao_nguoc(current_n)
            buoc += 1

        print(f"Số bước thực hiện: {buoc}")
        print(f"Giá trị palindrome thu được: {current_n}")

except ValueError:
    print("Lỗi: Đầu vào không phải là số nguyên hợp lệ.")
```

```python
#W4A9
from math import sqrt
n = int(input())
def check(x):
    cnt = [0] * 10
    while x:
        t = x % 10
        cnt[t] += 1
        x //= 10
        if cnt[t] > 1:
            return False
    return True

for i in range(1, int(sqrt(n))):
    if check(i * i):
        print(i * i, end = " ")
```

```python
#W4A10
n = int(input())
ans = 0
val = 0 

for x in range(1, n + 1):
    res = 1
    original = x
    while x != 1:
        if x % 2 == 0:
            x /= 2 
        else:
            x = x * 3 + 1
        res += 1
    if res > ans:
        ans = res
        val = original

print(val, ans)
```

```python
#W4A11
n = int(input("Nhập số nguyên dương n (< 10^6): "))
count = 0
i = 1

while i * i <= n:
    if n % i == 0:
        uoc1 = i
        uoc2 = n // i
        if uoc1 % 2 == 0:
            count += 1
        if uoc2 != uoc1 and uoc2 % 2 == 0:
            count += 1

    i += 1

print("Số lượng ước số chẵn của", n, "là:", count)
```

```python
#W4A13
def tong_uoc(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong
a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))
if tong_uoc(a) == b and tong_uoc(b) == a:
    print("true")
else:
    print("false")
```

```python
#W4A15
C = int(input("Nhập tổng số con: "))
Z = int(input("Nhập tổng số chân: "))

tim_thay = False
so_cho = 0
while so_cho <= C and not tim_thay:
    so_ga = C - so_cho
    tong_chan_kiem_tra = (so_ga * 2) + (so_cho * 4)

    if tong_chan_kiem_tra == Z:
        print(f"Số gà: {so_ga}")
        print(f"Số chó: {so_cho}")
        tim_thay = True

    so_cho += 1

if not tim_thay:
    print("invalid")
```
